2. Find the BusinessTypeID for "Restaurant/Cafe/Canteen" and return only the `BusinessTypeID` and `BusinessType` fields.

```python
# Find the BusinessTypeID for "Restaurant/Cafe/Canteen" and return only the BusinessTypeID and BusinessType fields
# Query for the BusinessTypeID "Restaurant/Cafe/Canteen"
query = {"BusinessType": "Restaurant/Cafe/Canteen"}

# Projection to return only the BusinessTypeID and BusinessType fields
projection = {"BusinessTypeID","BusinessType"}

# Find the matching documents with the specified query and projection
results = collection.find(query, projection)

# Iterate over the results and print each document
for document in results:
    pprint(document)
```

3. Update the new restaurant with the `BusinessTypeID` you found.

```python
# Update the new restaurant with the correct BusinessTypeID

# Specify the filter/query condition as an empty dictionary to match all documents
filter = {"BusinessName":"Penang Flavours"}

# Set the new value for the "BusinessTypeID" field
update = {"$set": {"BusinessTypeID": "1"}}

# Perform the update operation on all documents in the collection
result = collection.update_many(filter, update)

# Check if the update was successful
if result.modified_count > 0:
    print("BusinessTypeID field updated for query.")
else:
    print("No documents found or update failed.")
```

```python
# Change the data type from String to Decimal for latitude
import decimal

# Query for documents with the latitude field as a string
query = {"latitude": {"$type": "string"}}

# Find documents that match the query
documents = collection.find(query)

# Iterate over the documents and update the latitude field
for document in documents:
    # Convert the latitude value to decimal
    latitude_decimal = decimal.Decimal(document["latitude"])
    
    # Update the document with the new latitude value
    update = {"$set": {"latitude": latitude_decimal}}
    collection.update_one({"_id": document["_id"]}, update)

print("Data type conversion completed.")
```

```python
# Check that the coordinates are now numbers
# Query a document from the collection
document = collection.find_one()

# Access the geocode dictionary
geocode = document["geocode"]

# Check the data type of the latitude field
latitude_type = type(geocode["latitude"])
print("Data type of latitude field:", latitude_type)

# Check the data type of the longitude field
longitude_type = type(geocode["longitude"])
print("Data type of longitude field:", longitude_type)


```
